```python
def convolution(sequence1, sequence2):
    # Get the lengths
    len_seq1 = len(sequence1)
    len_seq2 = len(sequence2)
    
    # Prepare result list of appropriate length
    result = [0] * (len_seq1 + len_seq2 - 1)
    print(result)
    
    # perform convolution
    for i in range(len_seq1):
        for j in range(len_seq2):
            result[i+j] += sequence1[i] * sequence2[j]
    
    return result
```

```python
import matplotlib.pyplot as plt

def plot_sequence(sequence, title):
    plt.figure(figsize=(14, 4))
    plt.step(range(len(sequence)), sequence, where='post')
    plt.title(title)
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.show()

def tripleplot(inputSequence, kernel, output):
    plot_sequence(inputSequence, 'Input Sequence')
    plot_sequence(kernel, 'Kernel')
    plot_sequence(output, 'Output Sequence')
```

```python
# Smoothing
s = [0,0,0,0,0,0,0,0,4,4,4,4,0,0,0,0,0,0,0]
kernel= [0.3,0.3,0.3]
s_conv = convolution(s, kernel)
tripleplot(s,kernel,s_conv)
```
